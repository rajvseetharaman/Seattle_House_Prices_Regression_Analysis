```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
housing_data=pd.read_csv('Seattle 6 month home sales.csv')
```

```python
import scipy as sc
x=[]
x1=y=np.array(housing_data['SQUARE FEET'][:-20])
x2=np.array(housing_data['BEDS'][:-20])
x3=y=np.array(housing_data['BATHS'][:-20])
x4=y=np.array(housing_data['LOT SIZE'][:-20])
dataset_length=len(x1)
x0=[]
for i in range(dataset_length):
    x0.append(1)
x0=np.array(x0)
x.append(x0)
x.append(x1)
x.append(x2)
x.append(x3)
x.append(x4)
hx=np.array(x)
hx_transpose=np.matrix(hx)
hx=hx_transpose.transpose()
y=np.matrix(np.array(housing_data['PRICE'][:-20])).transpose()
w1=np.linalg.inv(hx_transpose*hx)*hx_transpose*y

testx=np.array(hx[-20:])
testy=np.array(y[-20:])
rmse=np.mean(np.array(testx*w1-testy)**2)**0.5
print(rmse)
plt.plot(testx,testy,marker='+',linestyle='none')
plt.show()
plt.plot(testx,testx*w1,marker='.',linestyle='none')
plt.show()





    
```
